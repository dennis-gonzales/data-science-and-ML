```python
# car_sales.dropna(inplace=True)
# car_sales.fillna(<DATA_TO_FILL>, inplace=True)

# `inplace=True` is an optional parameter used in pandas methods to modify the original DataFrame without creating a new copy of it. 
# When we set inplace=True, the modifications we make to the DataFrame persist and are saved in the original DataFrame.

# If `inplace=False` (default value), the methods return a new copy of the modified DataFrame, leaving the original DataFrame unmodified.
```

```python
# Creating column from Series
sold_to = pd.Series(['Dennis', 'Dennis', 'Dennis', 'Dennis', 'Dennis'])
car_sales["sold_to"] = sold_to
car_sales
```

```python
car_sales["sold_to"].fillna("Jasper", inplace=True) # fills NaN values and uses inplace
car_sales
```

```python
# Creating column from Python list

# Import random (to be set as odometer value)
import random

# Using list comprehention to populate list with random
odometer = [random.randint(1, 999998) for _ in range(len(car_sales))]
odometer

# Place the changes
car_sales["odometer_in_km"]  = odometer
car_sales
```

```python
car_sales["is_insured"] = car_sales["year"] > 2021
car_sales
```

```python
# car_sales.drop("is_licensed", axis=1, inplace=True) # dropping a column

# In pandas, the axis parameter is used to specify the dimension of the dataframe along which an operation should be performed.

# When axis=0, it refers to the rows of the dataframe.
# When axis=1, it refers to the columns of the dataframe.
```

```python
# Shuffle data frame

car_sales_shuffled = car_sales.sample(frac=1)
car_sales_shuffled
```

```python
# Unshuffle index
car_sales_shuffled.reset_index(drop=True, inplace=True) # drop=True Do not try to insert index into DataFrame columns
car_sales_shuffled
```

```python
# In Pandas, the .apply() method is used to apply a function to each element or row of a DataFrame.
# The lambda function is a small anonymous function that can take any number of arguments and returns a value based on the given expression.

car_sales["odometer_in_miles"] = car_sales["odometer_in_km"].apply(lambda x: x / 1.6)
car_sales
```
